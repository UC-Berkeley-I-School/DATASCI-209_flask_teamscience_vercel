```python
# dropping the duplicated columsn (suffix _y)
nih_merged2 = nih_merged.drop(columns=[col for col in nih_merged.columns if col.endswith('_y')])

# remove the '_x' suffix from remaining columns
nih_merged2.columns = nih_merged2.columns.str.replace('_x', '')
```

```python
date_cols = ['latest_disruption_date',
 'latest_restoration_date',
 'first_termination_date',
 'latest_termination_date',
 'first_reinstatement_date',
 'latest_reinstatement_date',
 'frozen_date',
 'unfrozen_date',
 'latest_outlay',
 'latest_renewal',
 'latest_budget_end',
 'project_start',
 'project_end',
 'project_start_date',
 'project_end_date',
 'award_notice_date',
 'budget_start', 'budget_end']
nih_merged2[date_cols] = nih_merged2[date_cols].apply(pd.to_datetime)
```
